#### Keep in mind that, despite being used for classification purposes, Logistic Regression is essentially an activation function imposed on Linear Regression in order to tailor to the needs of classification problems. This means we must not have categorical features.

To understand this better, think about how a linear regression model will react to the addition of some outliers in our data?

Surely, a linear model will depend heavily on the outliers and even though it might seem intuitive that a sample having a huge weight is probably going to be categorized as obese, but what our Linear Model will do is shift the fitted line and decrease its slope in order to minimize the SSE (including that for the outliers) and thus will end up changing the decision making boundary substantially.

```python
class LogisticRegressionImp:
    
    def __init__(self, learning_rate=0.001, epochs=5000):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.w = None
        self.b = None
    
    def __sigmoid(self, z):
        return (  1 / (1 + np.exp(-z))  )
    
    
    def __cost(self, X, y):
        cost_1 = -y*np.log(self.__hypothesis(X))
        cost_2 = -(1-y)*(np.log(1-self.__hypothesis(X)))
        return (1/self.m) * np.sum(cost_1 + cost_2) 
    
    
    def __hypothesis(self, X):
        return self.__sigmoid(X.dot(self.w) + self.b)
        
        
    def __update_weights(self, X, y):
        h = self.__hypothesis(X)
        dw = np.dot(X.T, (h-y))/self.m
        db = np.sum(h - y)/self.m
        self.w = self.w - self.learning_rate * (dw)
        self.b = self.b - self.learning_rate * (db)
    
    
    def fit(self, X, y):
        self.m, self.n = X.shape
        self.w = np.zeros((self.n))
        self.b = 0
        self.X = X
        self.y = y
        for i in range(self.epochs):
            self.__update_weights(self.X, self.y)
        return self
    
    
    def predict(self, X):
        h = self.__hypothesis(X)
        return np.where(h>=0.5, 1, 0)
    
    
    def accuracy(self, X_test, y_test):
        correct_pred = 0
        for i in range(len(X_test)):
            if self.predict(X_test[i, :]) == y_test[i]:
                correct_pred+=1
        return correct_pred/X_test.shape[0]
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
model_lr = LogisticRegressionImp(0.0002)
```

```python
from sklearn.linear_model import LogisticRegression
model_sk = LogisticRegression(max_iter=5000)
```

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=10)
```

```python
skf.get_n_splits()
```

```python
my_score = []
sk_score = []
for train_indx, test_indx in skf.split(X, y):
    X_train, X_test = X[train_indx], X[test_indx]
    y_train, y_test = y[train_indx], y[test_indx]
    model_lr.fit(X_train, y_train)
    model_sk.fit(X_train, y_train)
    my_score.append(model_lr.accuracy(X_test, y_test))
    sk_score.append(model_sk.score(X_test, y_test))    
```

```python
print("My score =", np.mean(my_score))
print("SkLearn's score =", np.mean(sk_score))
```
