```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Generating random data using np.random.multivariate_normal

```python
mean = np.array([4.0, 10.0])
cov = np.array([[1.0, 0.9], [0.9, 1.0]])
data = np.random.multivariate_normal(mean, cov, 8000)
```

```python
plt.scatter(data[:, 0], data[:, 1], marker = '.')
plt.show()
```

#### Shuffling the data to make the input random when it is fed in our model

```python
# shuffling the data for random splitting in the future even though this can be skipped as we already
# have random data
np.random.shuffle(data)
print(data[:5, :])
```

#### Splitting and Treating the data form training and testing

```python
X_train = data[:int(0.9*len(data)), 0]
X_test = data[-int(0.1*len(data)):, 0]
y_train = data[:int(0.9*len(data)), 1]
y_test = data[-int(0.1*len(data)):, 1]
```

```python
X_train = np.stack((np.ones(len(X_train)), X_train), axis=1)
X_test = np.stack((np.ones(len(X_test)), X_test), axis=1)
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
data = np.hstack((X_test, y_test))
data.shape
```

#### Making the necessary functions

```python
def hypothesis(X, theta):
#   (240, 2), (2, 1)
    return np.dot(X, theta)
```

```python
def cost(X, y, theta):
    h = hypothesis(X, theta)
    return np.sum((h-y)**2)
```

```python
def create_mini_batches(X, y, batch_size):
    mini_batches = []
    n_minibatches = X.shape[0] // batch_size
    i = 0
    for i in range(n_minibatches + 1):
        X_mini = X[i*batch_size:(i+1)*batch_size, :]
        y_mini = y[i*batch_size:(i+1)*batch_size, :]
        mini_batches.append((X_mini, y_mini))
    
    if X.shape[0] % batch_size != 0:
        X_mini = X[i * batch_size:data.shape[0]]
        y_mini = y[i * batch_size:data.shape[0]]
        mini_batches.append((X_mini, y_mini))
        
    return mini_batches
```

```python
def gradient(X, y, theta):
    h = hypothesis(X, theta)
    grad = np.dot(X.transpose(), (h - y))
    return grad
```

```python
def mini_batch_gradient_descent(X, y, learning_rate = 0.001, batch_size = 32):
    theta = np.zeros((X.shape[1], 1))
    error_list = []
    max_iters = 3
    for itr in range(max_iters):
        mini_batches = create_mini_batches(X, y, batch_size)
        for mini_batch in mini_batches:
            X_mini, y_mini = mini_batch
            theta = theta - learning_rate * gradient(X_mini, y_mini, theta)
            error_list.append(cost(X_mini, y_mini, theta))
  
    return theta, error_list
```

```python
theta, error_list = mini_batch_gradient_descent(X_train, y_train, 0.0023, 10)
```

#### Inspecting our result by finding the cost of the model

```python
print("Bias = ", theta[0])
print("Coefficients = ", theta[1:])
# visualising gradient descent
plt.plot(error_list)
plt.xlabel("epochs")
plt.ylabel("Cost")
plt.show()
```

#### Plotting the model against training and testing data

```python
plt.scatter(X_train[:, 1], y_train[:,], marker='x')
plt.plot(X_train[:, 1], hypothesis(X_train, theta), color='black')
print("Mean error on the training data =", cost(X_train, y_train, theta)/X_train.shape[0])
plt.show()
```

```python
plt.scatter(X_test[:, 1], y_test[:,], marker='x')
plt.plot(X_test[:, 1], hypothesis(X_test, theta), color='orange')
print("Mean error on the testing data =", cost(X_test, y_test, theta)/y_test.shape[0])
```
